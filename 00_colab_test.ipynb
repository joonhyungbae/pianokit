# 00. Colab Environment Test

This notebook verifies that the Colab runtime is working correctly:
1. **CPU** — numpy matrix multiplication
2. **GPU** — PyTorch CUDA availability and tensor operation
3. **Audio** — generate and play a sine wave in the browser

```python
# --- Cell 1: CPU Test ---
import time
import numpy as np

size = 2000
A = np.random.randn(size, size).astype(np.float32)
B = np.random.randn(size, size).astype(np.float32)

t0 = time.perf_counter()
C = A @ B
elapsed = time.perf_counter() - t0

print(f"CPU matrix multiply ({size}x{size}): {elapsed:.3f}s")
print(f"Result shape: {C.shape}, sample value: {C[0, 0]:.4f}")
print("\n=== CPU OK ===")
```

```python
# --- Cell 2: GPU Test ---
import torch

if torch.cuda.is_available():
    gpu_name = torch.cuda.get_device_name(0)
    print(f"GPU detected: {gpu_name}")
    print(f"CUDA version: {torch.version.cuda}")

    x = torch.randn(1000, 1000, device="cuda")
    y = torch.randn(1000, 1000, device="cuda")

    t0 = time.perf_counter()
    z = x @ y
    torch.cuda.synchronize()
    elapsed = time.perf_counter() - t0

    print(f"GPU matrix multiply (1000x1000): {elapsed:.4f}s")
    print("\n=== GPU OK ===")
else:
    print("GPU not available — CPU-only runtime.")
    print("Tip: Runtime > Change runtime type > T4 GPU")
    print("\n=== GPU SKIPPED (CPU-only) ===")
```

```python
# --- Cell 3: Audio Test ---
from IPython.display import Audio, display

sr = 22050
duration = 2.0
freq = 440.0  # A4

t = np.linspace(0, duration, int(sr * duration), endpoint=False)
wave = 0.5 * np.sin(2 * np.pi * freq * t)

print(f"Generated {duration}s sine wave at {freq}Hz (sample rate {sr})")
print("Press the play button below to verify audio output.")
display(Audio(wave, rate=sr))

print("\n=== Audio OK ===")
```
