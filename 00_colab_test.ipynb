이 노트북은 Colab이 제대로 작동하는지 확인합니다.  
각 셀을 위에서부터 순서대로 실행(▶ 버튼 또는 `Shift+Enter`)하세요.

| 단계 | 확인 항목 | 설명 |
|------|----------|------|
| 1 | **CPU** | 기본 연산이 되는지 확인 |
| 2 | **GPU** | 딥러닝용 그래픽카드가 연결되어 있는지 확인 |
| 3 | **오디오** | 소리가 재생되는지 확인 |

세 단계 모두 통과하면 다음 노트북으로 넘어가면 됩니다.

```python
# --- 1단계: CPU 테스트 ---
import time
import numpy as np

print("① CPU 테스트")
print("   컴퓨터의 기본 연산 능력을 확인합니다...\n")

size = 2000
A = np.random.randn(size, size).astype(np.float32)
B = np.random.randn(size, size).astype(np.float32)

t0 = time.perf_counter()
C = A @ B
elapsed = time.perf_counter() - t0

print(f"   {size}×{size} 행렬 곱셈 완료 — {elapsed:.2f}초 소요")
print(f"\n✅ CPU 정상 작동!")
```

```python
# --- 2단계: GPU 테스트 ---
import torch

print("② GPU 테스트")
print("   AI 모델 실행에 필요한 그래픽카드(GPU)를 확인합니다...\n")

if torch.cuda.is_available():
    gpu_name = torch.cuda.get_device_name(0)
    print(f"   GPU 발견: {gpu_name}")

    x = torch.randn(1000, 1000, device="cuda")
    y = torch.randn(1000, 1000, device="cuda")

    t0 = time.perf_counter()
    z = x @ y
    torch.cuda.synchronize()
    elapsed = time.perf_counter() - t0

    print(f"   GPU 행렬 곱셈 완료 — {elapsed:.4f}초 소요")
    print(f"\n✅ GPU 정상 작동!")
else:
    print("   ⚠️  GPU가 연결되어 있지 않습니다.")
    print()
    print("   해결 방법:")
    print("   1. 상단 메뉴에서 [런타임] → [런타임 유형 변경] 클릭")
    print("   2. 하드웨어 가속기를 'T4 GPU'로 선택")
    print("   3. 저장 후 이 셀을 다시 실행")
    print(f"\n⏭️  GPU 건너뜀 (CPU 전용 모드)")
```

```python
# --- 3단계: 오디오 테스트 ---
from IPython.display import Audio, display

print("③ 오디오 테스트")
print("   소리가 재생되는지 확인합니다...\n")

sr = 22050
duration = 2.0
freq = 440.0  # A4 (가온다 위의 라)

t = np.linspace(0, duration, int(sr * duration), endpoint=False)
wave = 0.5 * np.sin(2 * np.pi * freq * t)

print(f"   A4 (440Hz) 사인파 {duration}초 생성 완료")
print("   ▶ 아래 재생 버튼을 눌러 '라' 음이 들리는지 확인하세요.\n")
display(Audio(wave, rate=sr))

print("✅ 오디오 정상 작동!")
print()
print("=" * 40)
print("🎉 모든 테스트를 통과했습니다!")
print("   다음 노트북으로 넘어가세요.")
print("=" * 40)
```
